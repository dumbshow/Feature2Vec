```python
pivot_classes = ['hammer', 'dog', 'apple', 'sink', 'rose']
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_classes(word, max_classes):
    
    vectors = np.zeros((max_classes, 300))
    mat = cosine_similarity([model.wvector(word)], model.embedding_matrix.T)
    words = np.flip([model.id2concept[num] for num in np.argsort(mat[0,:])[max_classes:]])[:max_classes]
    for index, word in enumerate(words):
        vectors[index,:] = model.embedding_matrix.T[model.concept2id[word]]
        
    return words, vectors
```

```python
words, vectors = get_classes(pivot_classes[0], 5)
```

```python
for word in pivot_classes[1:]:
    twords, tvectors = get_classes(word, 5)
    for w in twords:
        words = np.append(words, w)
        
    vectors = np.vstack([vectors, tvectors])
```

```python
from collections import Counter 

counter = Counter([f[0] for concept in model.concepts for f in model.concept_features[concept]])
```

```python
top_features = [x[0] for x in counter.most_common()[:30]]
```
